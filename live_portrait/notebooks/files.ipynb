```python
# load txt and val file for hdtf
txt_file = Path("/data/prakash_lipsync/txt_files/iv_recording_v2/train.txt")
val_file = Path("/data/prakash_lipsync/txt_files/iv_recording_v2/val.txt")

# load the files
with open(txt_file, "r") as f:
    train_files = f.readlines()

with open(val_file, "r") as f:
    val_files = f.readlines()

# convert to dataframe
df_train = pd.DataFrame(train_files, columns=["img_id"])
df_val = pd.DataFrame(val_files, columns=["img_id"])

# combine the dataframes
df = pd.concat([df_train, df_val], ignore_index=True)
df.head()
```

```python
# remove the \n from the img_id
df["img_id"] = df["img_id"].apply(lambda x: x.strip())
df.head()
```

```python
# for each img_id get the total bins and frames
img_ids = df["img_id"].unique()
```

```python
from tqdm import tqdm

MAIN_PATH = "/data/prakash_lipsync/video_hallo_512"
mapping_dict = []

for img_id in tqdm(img_ids, desc="Processing images", total=len(img_ids)):
    # get the total bins and frames
    full_path = Path(f"{MAIN_PATH}/{img_id}")
    bins = list(full_path.glob("*"))
    for bin in bins:
        frames = list(bin.glob("*"))
        for frame in frames:
            mapping_dict.append(
                {
                    "clip_id": img_id,
                    "bin_id": str(bin).split("/")[-1],
                    "dataset": "iv_recording_v2",
                    "img_path": str(frame),
                }
            )

# convert to dataframe
df_mapping = pd.DataFrame(mapping_dict, columns=["clip_id", "bin_id", "dataset", "img_path"])
df_mapping.head()

# save to csv
df_mapping.to_csv("mapping.csv", index=False)
```

```python
# Method 1: Using numpy array_split (recommended)
# import numpy as np

# # Split into 8 parts
# df_splits = np.array_split(df_mapping, 8)

# # Save each partition
# for idx, split_df in enumerate(df_splits):
#     split_df.to_csv(f"partition_hdtf_{idx}.csv", index=False)

# Alternative Method 2: Using chunk size
chunk_size = len(df_mapping) // 40
for i in range(40):
    start_idx = i * chunk_size
    end_idx = start_idx + chunk_size if i < 39 else len(df_mapping)
    df_mapping.iloc[start_idx:end_idx].to_csv(f"partition_iv_recording_v2_{i}.csv", index=False)
```
